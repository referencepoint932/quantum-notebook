# 🔬 Quantum Superposition

## Understand the Hadamard gate (H)

The Hadamard gate creates *equal superposition* of ground states. It is defined as:
$$ H = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 & 1 \\
 1 & -1 \end{bmatrix} $$

### Quantum State

Any quantum state in a 2‑level system can be represented as a vector of complex amplitudes (phases):

$$ \psi = \begin{bmatrix} α \\ β \end{bmatrix} $$
$$ |α|² + |β|² = 1 $$

This can be re-expressed with ground state vectors (Dirac notation):
$$ |0\rangle = \begin{bmatrix} 1 \\ 0 \end{bmatrix} $$
$$ |1\rangle = \begin{bmatrix} 0 \\ 1 \end{bmatrix} $$
$$ |\psi\rangle = α\,|0\rangle + β\,|1\rangle $$

The Hadamard gate creates *equal superposition* of both ground states:
$$H\,|0\rangle \;=\; \frac{|0\rangle + |1\rangle}{\sqrt{2}}$$
$$H\,|1\rangle \;=\; \frac{|0\rangle - |1\rangle}{\sqrt{2}}$$

If we square any of those to obtain probabilities, we lose the minus sign and get 50/50.
Think of it as a **perfectly fair quantum coin flip**.

### Bloch Sphere

The Bloch sphere is a powerful tool for visualizing the state of a single qubit.  The north pole represents the ground state |0⟩, while the south pole represents the excited state |1⟩.  Any point on the surface corresponds to a superposition of these states, with the angles determining the relative amplitudes and phases.

```python
qc_ground = QuantumCircuit(2)
qc_ground.x(1)
qc_superposition = QuantumCircuit(2)
qc_superposition.h(0)
qc_superposition.x(1)
qc_superposition.h(1)
state = Statevector.from_instruction(qc_ground)
plot_bloch_multivector(state, title="Ground states", font_size=16)
```

```python
qc_superposition = QuantumCircuit(2)
qc_superposition.h(0)
qc_superposition.x(1)
qc_superposition.h(1)
state = Statevector.from_instruction(qc_superposition)
plot_bloch_multivector(state, title="H|0> and H|1> Superpositions", font_size=16)
```
